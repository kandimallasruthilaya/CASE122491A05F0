```python
import matplotlib.pyplot as rajamouli
movies=['stu no.1','yamadhonga','maryadharamana','eega','rrr']
amount=[12,40,30,125,1200]
rajamouli.title('balaya movies stats')
rajamouli.pie(amount,labels=movies,autopct='%1.1f%%')
c=rajamouli.Circle((0,0),0.6,color="white")
rajamouli.gca().add_artist(c)
rajamouli.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
sruthi=pd.read_csv('energy.csv')
print(sruthi)
#line chart for ap and karnataka
srh=sruthi[sruthi['State']=='Andhra']
rcb=sruthi[sruthi['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='red')
diag[1].plot(rcb['Date'],rcb['Consumption'],color='green')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
sumanth=pd.read_csv('energy.csv')
print(sumanth)

#line chart
srh=sumanth[sumanth['State']=='Andhra']
rcb=sumanth[sumanth['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='pink')
diag[1].plot(rcb['Date'],rcb['Consumption'],color='skyblue')

srh=sumanth[sumanth['State']=='Telengana']
rcb=sumanth[sumanth['State']=='Tripura']
fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'],color='lightgreen')
diag[1].bar(rcb['Date'],rcb['Consumption'],color='lightyellow')

sruthi='Kerala'
laya='Delhi'
tdp=[sruthi,laya]

srh=sumanth[sumanth['State']=='Kerala']['Consumption'].sum()
rcb=sumanth[sumanth['State']=='Delhi']['Consumption'].sum()
ysr=[srh,rcb]

fig,diag=plt.subplots(2)
diag[0].pie(ysr,labels=tdp)
```
